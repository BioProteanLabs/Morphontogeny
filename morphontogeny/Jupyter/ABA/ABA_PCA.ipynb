# PCA on ABA
**With different normalizations**:
* Including -1, scaled (std)
* Including -1, L2 normalized
* -1 to 0, scaled
* -1 to 0, L2 normalized

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
# Including -1s, and scaled
X = np.memmap('/data/bioprotean/ABA/MEMMAP/genes_list/finalgenes_std.mymemmap',\
dtype='float32', mode='r', shape=(159326,2941))

pca = PCA(n_components = 5)
components = pca.fit_transform(X)

#Visualizing one slice
c = components[:,0]
vol = c.reshape(67,58,41)
s = vol[20,:,:]
plt.imshow(s)
```

```python
# Including -1s, and L2 normalized
X = np.memmap('/data/bioprotean/ABA/MEMMAP/genes_list/finalgenes_L2.mymemmap',\
dtype='float32', mode='r', shape=(159326,2941))

pca = PCA(n_components = 5)
components = pca.fit_transform(X)

#Visualizing one slice
c = components[:,0]
vol = c.reshape(67,58,41)
s = vol[20,:,:]
plt.imshow(s)
```

```python
# -1s to 0s, and L2 normalized
X = np.memmap('/data/bioprotean/ABA/MEMMAP/genes_list/finalgenes_pos_L2.mymemmap',\
dtype='float32', mode='r', shape=(159326,2941))

pca = PCA(n_components = 5)
components = pca.fit_transform(X)

#Visualizing one slice
c = components[:,0]
vol = c.reshape(67,58,41)
s = vol[20,:,:]
plt.imshow(s)
```

```python
# -1s to 0s, and scaled
X = np.memmap('/data/bioprotean/ABA/MEMMAP/genes_list/finalgenes_pos_std.mymemmap',\
dtype='float32', mode='r', shape=(159326,2941))

pca = PCA(n_components = 5)
components = pca.fit_transform(X)

#Visualizing one slice
c = components[:,0]
vol = c.reshape(67,58,41)
s = vol[20,:,:]
plt.imshow(s)
```

# Seems like for PCA, -1s to 0s and the Standard Scaling works the best.

```python
# Running PCA for %90 variance
# -1s to 0s, and scaled
X = np.memmap('/data/bioprotean/ABA/MEMMAP/genes_list/finalgenes_pos_std.mymemmap',\
dtype='float32', mode='r', shape=(159326,2941))

variance = 0.9
pca = PCA(variance)
components = pca.fit_transform(X)

print('{} components for the variance of {}.'.format(pca.n_components_ , variance))

# Saving the components
np.save('/data/bioprotean/ABA/PCA/PC_90v.npy', components)
```
