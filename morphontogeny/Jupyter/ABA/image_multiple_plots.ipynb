```python
import numpy as np
import nibabel as nib
import pandas as pd
from skimage import filters
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading the ICs and mixing matrix
S_ = np.load('/data/bioprotean/SVD/ICA/IC_30/IC_S_30.npy')
A_ = np.load('/data/bioprotean/SVD/ICA/IC_30/IC_A_30.npy')
```

```python
# Loading the reconstructed data of PCA and K-means
clusters_file = '/data/bioprotean/SVD/clusters/complete/fixed_header/nclusters_fixed40.nii'
clusters_img = nib.load(clusters_file)
clusters_array = np.array(clusters_img.dataobj)
```

```python
# Choosing the component and making it a 3D array
component = S_[:,7].reshape(67,58,41)

# Visualizing 6 of its slices, the clusters, and the histograms
fig, axs = plt.subplots(6,3, figsize=(10,25))

# Defining the axes
axs[0,0].imshow(component[5,:,:])
axs[0,1].imshow(clusters_array[5,:,:])
axs[0,2].hist(component[5,:,:])


axs[1,0].imshow(component[15,:,:])
axs[1,1].imshow(clusters_array[15,:,:])
axs[1,2].hist(component[15,:,:])

axs[2,0].imshow(component[25,:,:])
axs[2,1].imshow(clusters_array[25,:,:])
axs[2,2].hist(component[25,:,:])

axs[3,0].imshow(component[35,:,:])
axs[3,1].imshow(clusters_array[35,:,:])
axs[3,2].hist(component[35,:,:])

axs[4,0].imshow(component[45,:,:])
axs[4,1].imshow(clusters_array[45,:,:])
axs[4,2].hist(component[45,:,:])

axs[5,0].imshow(component[55,:,:])
axs[5,1].imshow(clusters_array[55,:,:])
axs[5,2].hist(component[55,:,:])

plt.show()
```

```python
# Loading the genes list
genes_df = pd.read_csv('/data/bioprotean/SVD/genes_list/final_list.csv')
```

```python
# Taking the mixing values of genes for a slice
mixing_slice = A_[:,7]
mixing_df = pd.DataFrame(mixing_slice, columns=['mixing'])
```

```python
# Loading the genes list
genes_df = pd.read_csv('/data/bioprotean/SVD/genes_list/final_list.csv')
output_df = pd.concat([genes_df, mixing_df], axis=1)
```

```python
# Sorting the dataframe by mixing values (descending)
output_sorted_df = output_df.sort_values('mixing', ascending=False)
```

```python
# Top 10 genes list
top10_geneID_list = output_sorted_df.ID[0:9].tolist()
top10_geneID_list
```

```python
# Loading top hit genes
gene1_file = '/data/bioprotean/ABA/ISH/'+str(top10_geneID_list[2])+'.csv'
gene1_2D = np.genfromtxt(gene1_file, delimiter=',', skip_header=1, usecols=range(1,2379))
gene1_3D = gene1_2D.reshape((67,58,41))
```
