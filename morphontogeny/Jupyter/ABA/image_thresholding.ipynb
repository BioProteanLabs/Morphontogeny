# Thresholding images
Trying functions from https://scikit-image.org/docs/dev/api/skimage.filters.html

```python
import numpy as np
from skimage import filters
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading the ICs and mixing matrix
S_ = np.load('/data/bioprotean/SVD/ICA/IC_30/IC_S_30.npy')
A_ = np.load('/data/bioprotean/SVD/ICA/IC_30/IC_A_30.npy')
```

```python
# Visualizing a slice
component = S_[:,20].reshape(67,58,41)
slice = component[20,:,:]
plt.imshow(slice)
```

```python
# Thresholding: all
fig, ax = filters.try_all_threshold(slice, figsize=(6, 10), verbose=False)
```

```python
# Plotting the histogram of the slice
n, bins, patches = plt.hist(slice, bins=255, facecolor='blue', alpha=0.5)
plt.show()
```

```python
# Thresholding the slice based on the histogram
thresh_hist = -0.004
image_binary = slice > thresh_hist

# Visualizing the original image and the threshold horizontally
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(slice)
ax2.imshow(image_binary)
plt.show()
```
