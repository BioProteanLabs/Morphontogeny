```python
# Visualizing with Seaborn
sns.set(style = "darkgrid")
a_plot = sns.lineplot(x = score_df['K'], y = score_df['AMI'], dashes = False, palette=['black'])
b_plot = sns.lineplot(x = score_df['K'], y = score_df['Silhouette'], dashes = False, palette=['green'])
a_plot.set(ylim = (0, 0.7))
```
