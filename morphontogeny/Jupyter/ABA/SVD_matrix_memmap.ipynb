```python
import pandas as pd
import numpy as np
```

```python
# Loading the list_top_genes.csv which is downloaded from the Allen's clustering paper for top genes
list = pd.read_csv("/home/mabbasi6/morphometry/correlation/SVD/list_top_genes.csv", delimiter=',')
list_top = list[list['Used in analysis?']==1]
```

```python
# Loading a sample gene file to get the dimensions
gene = np.genfromtxt('/home/mabbasi6/morphometry/correlation/ish/70927828.csv', delimiter=',', skip_header=1, usecols=range(1,2379)).flatten()
```

```python
# Matrix is huge, so making a memmap to store the ISH data in one file
topgenes_mmp = np.memmap('/home/mabbasi6/morphometry/correlation/SVD/genes_mmp.mymemmap', dtype='float32', mode='w+', shape=(list_top.shape[0],gene.shape[0]))
```

```python
# First attemp to make a huge matrix of ISH data for top genes (but some of these genes are removed from dataset)
# Genes stored in rows, voxels as columns
for i in range (list_top.shape[0]):
    try:
        topgenes_mmp[i,:] = np.genfromtxt('/home/mabbasi6/morphometry/correlation/ish/'+list_top['ID'][i]+'.csv', \
                                    delimiter=',', skip_header=1, usecols=range(1,2379)).flatten()
    except:
        pass
```

```python
# Loading the dataframe of removed genes from the top ones
list_removed = pd.read_csv('/home/mabbasi6/morphometry/correlation/SVD/removed_genes.csv', delimiter=',')
list_removed
```

```python
# Making a memmap of the size of final dataset (top genes - removed genes)
finalgenes_mmp = np.memmap('/home/mabbasi6/morphometry/correlation/SVD/finalgenes_mmp.mymemmap', dtype='float32', \
                           mode='w+', shape=(list_top.shape[0]-list_removed.shape[0],gene.shape[0]))
```
