```python
'''
Defining a score to measure the goodness of the fit to the Neuroanatomy
Weighted average of max overlap ratios with weights being the size of the brain region
'''
# Computing the weight of each brain region
def region_weight(ant_vec, ant_id):
    '''
    Computes the weight of each brain region as weight = size of region / size of brain
    
    Args:
        ant_vec (vector): The vector of annotations
        ant_id (float): ID of the brain region
        
    Returns:
        weight (float): weight of the region
    '''
    # Total number of non-zero voxels
    n_total = ant_vec[ant_vec!=0].shape[0]
    
    # Number of voxels with brain region ID
    n_region = ant_vec[ant_vec == ant_id].shape[0]
    
    weight = n_region / n_total
    return weight
```

```python
def max_overlap_byregion(cluster_vec, ant_vec, ant_id):
    '''
    Takes 2 labeled vectors (anatomy and clusters for example).
    Computes the overlap of a brain region with the clusters and
    Returns the highest overlap ratio and the cluster ID of that.
    
    Args:
        cluster_vec (vec): 1D vector of the clusters
        ant_vec (vec): 1D vector of annotations
        ant_id: ID of brain region in the annotation file
        
    Returns:
        [max_ratio, max_id]
        max_ratio (float): Ratio of maximum overlap
        max_id (float): ID of the cluster that has maximum overlap
    '''
    
#     For masking:
#     Where annotations vector equals an annotation ID,
#     The cluster labels in the voxel will be kept untouched.
#     Otherwise, -1 is replaced.
#     This way, counting the ratio of overlap will be very simple.
    mask = np.where(ant_vec == ant_id, cluster_vec, -1)
    
    # Getting list of overlapping cluster IDs and their counts
    clusters_list, counts_list = np.unique(mask[mask != -1], return_counts=True)
    
    # Count of overall voxels in that brain region
    n_voxels = np.sum(counts_list)
    
    # Finding the maximum count and the corresponding cluster label
    max_count = np.amax(counts_list)
    max_index = np.argmax(counts_list)
    max_id = clusters_list[max_index]
    
    # Computing the ratio
    max_ratio = max_count / n_voxels
    
    return max_ratio, max_id
```

```python
'''
To penalize the huge size of clusters, now this new analysis would find the cluster
that has maxmimum overlap with the brain region and returns the ratio of overlap
with the CLUSTER.
'''
```

```python
def max_overlap_bycluster(cluster_vec, ant_vec, ant_id):
    '''
    Takes 2 labeled vectors (anatomy and clusters for example).
    Computes the overlap of a brain region with the clusters and
    returns the highest overlap ratio (with CLUSTER)
    and the cluster ID of that.
    
    Args:
        cluster_vec (vec): 1D vector of the clusters
        ant_vec (vec): 1D vector of annotations
        ant_id: ID of brain region in the annotation file
        
    Returns:
        [max_ratio, max_id]
        max_ratio (float): Ratio of maximum overlap
        max_id (float): ID of the cluster that has maximum overlap
    '''

#     For masking:
#     Where annotations vector equals an annotation ID,
#     The cluster labels in the voxel will be kept untouched.
#     Otherwise, -1 is replaced.
#     This way, counting the ratio of overlap will be very simple.
    mask = np.where(ant_vec == ant_id, cluster_vec, -1)
    
    # Getting the list of overlapping cluster IDs and their counts (as np vector)
    clusters_list, counts_list = np.unique(mask[mask != -1], return_counts=True)
    
    # Finding the maximum count and the corresponding cluster label
    max_count = np.amax(counts_list)
    max_index = np.argmax(counts_list)
    max_id = clusters_list[max_index]
    
    # Count of all voxels in that cluster
    n_voxels = cluster_vec[cluster_vec == max_id].shape[0]
    
    # Computing the ratio
    max_ratio = max_count / n_voxels
    
    return max_ratio, max_id
```
