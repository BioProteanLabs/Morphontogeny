# Figures of clustering performance metrics

```python
%run overlap_functions.ipynb
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# For PCA + K-means clusters

```python
# Loading the Adjusted Mutual Information score
score_path = '/data/bioprotean/ABA/PCA/80_variance/PC_80v_score.csv'
score_df = pd.read_csv(score_path, index_col = 'Unnamed: 0')
score_df.head()
```

```python
# Visualizing with Seaborn
sns.set(style = "darkgrid")
a_plot = sns.lineplot(x = score_df['K'], y = score_df['AMI'], dashes = False, palette=['black'])
b_plot = sns.lineplot(x = score_df['K'], y = score_df['Silhouette'], dashes = False, palette=['green'])
a_plot.set(ylim = (0, 0.7))
```

```python
# Visualizing with Matplotlib
plt.plot(score_df['K'], score_df['AMI'], color='black')
plt.plot(score_df['K'], score_df['Silhouette'], color='blue')
plt.grid(True)
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("Number of K-means clusters")
plt.ylabel("Clustering performance metric")
plt.legend(['AMI score', 'Silhouette score'], loc='lower right')
plt.ylim(0, 0.7)
# plt.title("Maximum overlap")

# To load the display window
# plt.show()
plt.savefig('/data/bioprotean/ABA/PCA/80_variance/score_fig.png')
```
