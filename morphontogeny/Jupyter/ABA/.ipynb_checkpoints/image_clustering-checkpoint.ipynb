```python
import numpy as np
import skimage
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading the ICs and mixing matrix
S_ = np.load('/data/bioprotean/SVD/ICA/IC_30/IC_S_30.npy')
A_ = np.load('/data/bioprotean/SVD/ICA/IC_30/IC_A_30.npy')
```

```python
# Visualizing a slice
component = S_[:,16].reshape(67,58,41)
slice = component[27,:,:]
plt.imshow(slice)
```

```python
# Plotting the histogram of the slice
n, bins, patches = plt.hist(slice, bins=255, facecolor='blue', alpha=0.5)
plt.show()
```

```python
# Seeing which bin each data point belongs to
data_bins = np.digitize(slice, bins=bins)
data_bins.shape
```

```python
# Making a 2D array of data and the corresponding bins
slice_1D = slice.reshape(-1)
data_bins_1D = data_bins.reshape(-1)
data_2D = np.column_stack((slice_1D, data_bins_1D))
```

```python
# Clustering the voxels
kmeans = KMeans(init='k-means++', n_clusters=3, n_init=50)
kmeans.fit_predict(data_2D)
kmeans_labels = kmeans.labels_
labels_slice = kmeans_labels.reshape((58,41))

# Visualizing the labels
fig_labels, (ax1_labels, ax2_labels) = plt.subplots(1,2)
ax1_labels.imshow(slice)
ax2_labels.imshow(labels_slice)
plt.show()
```

```python
# Thresholding the slice based on the histogram
thresh_hist = 0.004
image_binary = slice > thresh_hist

# Visualizing the original image and the threshold horizontally
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(slice)
ax2.imshow(image_binary)
plt.show()
```

```python
# Clustering the vector
kmeans2 = KMeans(init='k-means++', n_clusters=3, n_init=50)
kmeans2.fit_predict(slice.reshape(-1,1))
kmeans2_labels = kmeans2.labels_
print(slice.reshape(-1,1).shape)
labels2_slice = kmeans2_labels.reshape((58,41))

# Visualizing the labels
fig2_labels, (ax3_labels, ax4_labels) = plt.subplots(1,2)
ax3_labels.imshow(slice)
ax4_labels.imshow(labels2_slice)
plt.show()
```
