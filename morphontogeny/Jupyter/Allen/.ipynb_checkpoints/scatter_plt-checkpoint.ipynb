This is to scatter plot gene vs gene

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Loading both genes
load_ish1 = np.genfromtxt('/home/mabbasi6/morphometry/correlation/data/79556644.csv', \
                             delimiter=',', skip_header=1, usecols=range(1,2379)).flatten()
load_ish2 = np.genfromtxt('/home/mabbasi6/morphometry/correlation/data/69816709.csv', \
                             delimiter=',', skip_header=1, usecols=range(1,2379)).flatten()

# Replacing -1 values with NaN in order to avoid miscalculations in correlation analysis
load_ish1 [load_ish1 < 0] = np.nan
load_ish2 [load_ish2 < 0] = np.nan

# Making pandas Dataframe
inp1 = pd.DataFrame(load_ish1, columns=['ISH1'])
inp2 = pd.DataFrame(load_ish2, columns=['ISH2'])
df = pd.concat ([inp1,inp2], axis=1).dropna()

plt.scatter(df['ISH1'], df['ISH2'], marker='o')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

load_jac = np.loadtxt ('/home/mabbasi6/morphometry/correlation/flatten/ttest.csv', delimiter=',')
load_ish = np.genfromtxt('/home/mabbasi6/morphometry/correlation/data/69816709.csv', \
                             delimiter=',', skip_header=1, usecols=range(1,2379)).flatten()

load_ish [load_ish < 0] = np.nan
mask = np.heaviside (load_ish, 1)
masked_jac = np.multiply (mask, load_jac)

inp1 = pd.DataFrame(load_ish, columns=['ISH'])
inp2 = pd.DataFrame(masked_jac, columns=['JAC'])
df = pd.concat ([inp1,inp2], axis=1).dropna()

plt.scatter(df['ISH'], df['JAC'], marker='o')
```
