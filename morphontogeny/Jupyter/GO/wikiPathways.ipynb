```python
import numpy as np
import pandas as pd
```

```python
wiki = {}
with open('/data/bioprotean/GO/Chris/c2.cp.wikipathways.v7.2.entrez.gmt','r') as inFile:
    while 1:
        inLine = inFile.readline()
        if not inLine:
            break
        splitUp = inLine.strip().split('\t')
        wikiName = splitUp.pop(0)
        url = splitUp.pop(0)
        wiki[wikiName] = splitUp
```

```python
wikiDf = pd.DataFrame.from_dict(wiki, orient = 'index')
wikiDf = wikiDf.reset_index()
print(wikiDf.shape)
wikiDf.head()
```

```python
wikiDfMlt = wikiDf.melt(id_vars = 'index')
wikiDfMlt = wikiDfMlt[~pd.isnull(wikiDfMlt.value)]
print(wikiDfMlt.shape)
wikiDfMlt.head()
# (26619, 3) - wiki pathways; 587 unique pathways
```

```python
wikiDfMlt_2 = wikiDfMlt.sort_values(['index', 'variable'])
print(wikiDfMlt_2.shape)
wikiDfMlt_2.head()
```

```python
wikiDf_rename = wikiDfMlt_2.rename(columns={"index": "pathway", "value": "entrez"})
```

```python
wikiDf_rename.to_csv('/data/bioprotean/GO/Chris/wikiPathwaysDf.csv')
```
