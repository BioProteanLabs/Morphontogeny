# Finding receptor genes

```python
import numpy as np
import pandas as pd
import mygene
mg = mygene.MyGeneInfo()
```

```python
topgenes_df = pd.read_csv('/data/bioprotean/GO/top_genes.csv')
```

```python
topgenes_df.head()
```

```python
for i in range(topgenes_df.shape[0]):
    try:
        topgenes_df.gName[i] = mg.getgene(topgenes_df['Entrez ID'][i], fields='name')['name']
    except:
        continue
```

```python
topgenes_df
```

```python
topgenes_df.to_csv('/data/bioprotean/GO/topgeneswithnames.csv')
```

```python
topgenes_df[topgenes_df.gName.str.contains('receptor')].to_csv('/data/bioprotean/GO/receptors_only.csv')
```
