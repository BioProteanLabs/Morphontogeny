```python
import pandas as pd
import numpy as np
import nibabel as nib
import os
```

```python
internal_cap = pd.read_csv('/data/bioprotean/Naren/internal_capsule/6_internalcapsule.csv')
```

```python
for i in range (1,51):
    ic_column = internal_cap[str(i)].to_numpy()
    ic_column_rs = ic_column.reshape(67,41,58)
    img = nib.Nifti1Image(ic_column_rs, np.eye(4))
    img.header.get_xyzt_units()
    img.to_filename(os.path.join('build','/data/bioprotean/Naren/internal_capsule/ic_cluster_'+str(i)+'.nii'))
```

```python
%%bash
module load afni/20.1.11
allen=/data/bioprotean/Naren/internal_capsule/allen_50.nii
for i in {1..50}
do
    input=/data/bioprotean/Naren/internal_capsule/ic_cluster_$i.nii
    3drefit \
        -xorigin 0 -yorigin 0 -zorigin 0 \
        -orient ASR \
        -space ORIG \
        -view orig \
        $input
    3drefit -xyzscale 0.2 $input
    @Align_Centers \
        -base $allen \
        -dset $input \
        -overwrite
done
```
