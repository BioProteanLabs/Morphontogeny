We have our images warped to Henkelman template and the Jacobian and the ttest was calculated
in the Henkelman space.
This code warps Henkelman to Allen brain, and then applies same warps to the previous Jacobian and ttest data 
to bring them into Allen Brain space. The warp comprises of affine and nonlinear.

```python
%%bash
module load afni/20.1.11
cd /home/mabbasi6/morphometry/Allen/Dorr/
3dinfo -d3 masked+orig
```

```python
%%bash
module load afni/20.1.11
cd /home/mabbasi6/morphometry/Allen/Henkelman/
3dresample -dxyz 0.075 0.075 0.075 -input henkelman+orig -prefix henkel_75
```

```python
%%bash
module load afni/20.1.11
cd /home/mabbasi6/morphometry/Allen/Dorr/
3dAllineate -base allen_75+orig -source RsMDorr+orig -prefix henkel_aff -1Dmatrix_save affine -cmass -mi -twopass -automask -num_rtb 2000 -warp affine_general
```

```python
%%bash
module load afni/20.1.11
cd /home/mabbasi6/morphometry/Allen/Dorr/
auto_warp.py -base allen_75+orig -input henkel_aff+orig -suffix henkel_warp
```
