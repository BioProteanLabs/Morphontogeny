# Making average brain
Pipeline for getting average of 2 brains. Steps:
    1. Copy the datasets in the folder.
    2. Define the directory, input1 and input 2.
It will align the centers, resample, affine register, warp, and save the average brain as 'avgAB_qwpm'.
<br>_No space between variable and '=' sign_

```python
%%bash
module load afni/20.1.11
cd /home/mabbasi6/morphometry/MKO_Morphometry/1234/
input1='/home/mabbasi6/morphometry/MKO_Morphometry/1234/12+orig'
input2='/home/mabbasi6/morphometry/MKO_Morphometry/1234/34+orig'
3dcopy $input1 A
3dcopy $input2 B
@Align_Centers -base A+orig -dset B+orig -prefix B_ac
3dresample -master A+orig -input B_ac+orig -prefix B_acres
3dAllineate -base A+orig -source B_acres+orig -cmass -prefix B_aff
3dQwarp -base A+orig -source B_aff+orig -plusminus -pmNAMES Top Bottom -prefix avgAB
3dcalc -a avgAB_Top+orig -b avgAB_Bottom+orig -expr '(a+b)/2' -prefix avgAB_qwpm
```
